## Business Problem

An insurance company needs to identify the key features of a home that influence its market value. This information is crucial when estimating potential payouts in the event of a total loss. While the company understands that predicting the exact market value of a home based on its features is challenging, it aims to develop a model that can highlight the most significant features affecting home values.

#### The company's pain points:
The primary pain point for the insurance company is the complexity and variability of factors that determine a home's market value. While some features like location, size, and condition are known to impact value, there are potentially many other factors that could be influential. The company needs a way to identify and quantify the most important features to consider when estimating potential payouts.

#### The analysis question:
The main question we sought to answer through data analysis was: "What features of a home most significantly impact its market value?" This question was chosen because understanding these features can help the insurance company make more informed decisions when estimating potential payouts.

#### Why was this question chosen/important:
This question is important from a business perspective because it directly impacts the company's risk management and financial planning. By understanding the key features that influence home values, the company can make more informed estimates of potential payouts, helping it to manage risk and plan for financial contingencies. Furthermore, by identifying the most significant features, the company can focus its resources on collecting and analyzing data related to these features, leading to more efficient operations.

#### Source of data:
The data used for this project is the King County House Sales dataset. The dataset contains house sale prices for King County in Washington State, which includes Seattle. It includes homes sold between May 2014 and May 2015.

#### The data's relation to the analysis:
The data was sourced from Kaggle, a popular platform for open datasets. The dataset directly relates to our data analysis question as it contains information about various features of homes and their corresponding sale prices. By analyzing this data, we can identify the features that most significantly impact a home's market value.

#### What the data represents:
The dataset represents different houses in King County. Each row in the dataset corresponds to a unique house sale and includes information about various features of the house at the time of the sale. The sample includes a wide range of homes with varying features and sale prices.

#### The dataset includes the following variables:

id: Unique identifier for each home sold
date: Date of the home sale
price: Price of each home sold (our target variable)
bedrooms: Number of bedrooms
bathrooms: Number of bathrooms
sqft_living: Square footage of the home's interior living space
sqft_lot: Square footage of the land space
floors: Number of floors
waterfront: Whether the home has a view to a waterfront
view: An index from 0 to 4 of how good the view of the property was
condition: An index from 1 to 5 on the condition of the apartment,
grade: An index from 1 to 13, where 1-3 falls short of building construction and design, 7 has an average level of construction and design, and 11-13 have a high quality level of construction and design.
sqft_above: The square footage of the interior housing space that is above ground level
sqft_basement: The square footage of the interior housing space that is below ground level
yr_built: The year the house was initially built
yr_renovated: The year of the house’s last renovation
zipcode: What zipcode area the house is in
lat: Lattitude
long: Longitude
sqft_living15: The square footage of interior housing living space for the nearest 15 neighbors
sqft_lot15: The square footage of the land lots of the nearest 15 neighbors

#### Target Variable:
The target variable for this analysis is price, which represents the sale price of each home. The goal is to predict this variable based on the other features in the dataset.

#### Properties of Variables Used:
We used the following variables in our analysis: condition, total_sqft (a new variable we created by adding sqft_living, sqft_above, and sqft_basement), view, and grade. These variables were chosen based on their potential to influence a home's market value. They are all numerical variables, although condition, view, and grade are categorical variables which were encoded with a specific range of possible values for easier analysis.

```python
# Making a copy of the df before cleaning as a contingency
df_clean = df.copy()
df_clean.head()
```

```python
#Dropping the yr_renovated column from df_clean
df_clean = df_clean.drop('yr_renovated', axis=1)
```

```python
#Dealing with missing values in waterfront column
df_clean['waterfront'] = df_clean['waterfront'].fillna('NO')
```

```python
# Filling in missing values in view with the mode as the data is quite skewed as observed in the bar chart
mode_view = df_clean['view'].mode()[0]
df_clean['view'] = df_clean['view'].fillna(mode_view)
```

```python
df_clean.isnull().sum()
```

```python
#Cleaning the sqft_basement column
# Replacing '?' with '0'
df_clean['sqft_basement'] = df_clean['sqft_basement'].replace('?', '0')

# Convert the column to a numeric type
df_clean['sqft_basement'] = pd.to_numeric(df_clean['sqft_basement'])
```

```python
print(df_clean['sqft_basement'].describe())
```

```python
# Convert 'date' column to datetime type
df_clean['date'] = pd.to_datetime(df_clean['date'])
```

```python
# Checking new shape of our df after cleaning 
print('Shape after dropping:', df_clean.shape)
```

```python
#Encoding categorical variables 
# Condition
df_clean['condition'] = df_clean['condition'].map({'Average': 2, 'Good': 3, 'Very Good': 4, 'Fair': 1, 'Poor': 0})

# View
df_clean['view'] = df_clean['view'].map({'NONE': 1, 'AVERAGE': 2, 'GOOD': 3, 'FAIR': 4, 'EXCELLENT': 5, 'Unknown': 0})

# Waterfront
df_clean['waterfront'] = df_clean['waterfront'].map({'NO': 1, 'YES': 2, 'Unknown':0})

# Grade
df_clean['grade'] = df_clean['grade'].map({ '13 Mansion': 11 ,'12 Luxury': 10 , '11 Excellent': 9 , '10 Very Good': 8 , '9 Better': 7 , '8 Good': 6, '7 Average': 5 , '6 Low Average': 4 , '5 Fair': 3 , '4 Low': 2 , '3 Poor': 1 })

```

```python
print(df_clean.dtypes)
```

```python
# Identifying and dealing with outliers
num_cols = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'sqft_above', 'yr_built', 'lat', 'long', 'sqft_living15', 'sqft_lot15', 'sqft_basement']

for col in num_cols:
    Q1 = df_clean[col].quantile(0.25)
    Q3 = df_clean[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df_clean = df_clean[(df_clean[col] >= lower_bound) & (df_clean[col] <= upper_bound)]

```

```python
df_clean.info()
```

#### Some more EDA after cleaning:

```python
# Calculating the correlation
# Selecting only the numeric columns (excluding 'price')
numeric_columns = df_clean.select_dtypes(include=[np.number]).columns


correlation_with_price = df_clean[numeric_columns].corr()['price'].sort_values(ascending=False)
plt.figure(figsize=(10,8))
correlation_with_price.drop('price').plot(kind='bar')
plt.title('Correlation with Price')
plt.show()
```

After encoding our categorical variables, we find that there is a high positive correlation between grade and price.   

```python
#Creating a heatmap with correlation coefficients of df_clean
numeric_dfclean = df_clean.select_dtypes(include=[np.number])
correlation_matrix = numeric_dfclean.corr().round(2)
plt.figure(figsize=(12, 8))
sns.heatmap(data=correlation_matrix, annot=True)
```

From this heatmap we infer that the most positive highly correlated variables with house prices are: sqft_living, grade, lat, sqft_living15, and sqft above.

#### Creating a new feature for the total square footage 
sqft_living, sqft_above and sqft_basement are all highly correlated with the price, they are also positively correlated with each other. Total square feet will take all these variables into consideration for easier comparison with the target variable.

```python
# Create a new feature 'total_sqft'
df_clean['total_sqft'] = df_clean['sqft_living'] + df_clean['sqft_above'] + df_clean['sqft_basement']
df_clean.head()
```

### Simple Linear Regression

```python
#Calculating the correlation coefficient of our new feature and price
correlation = df_clean['total_sqft'].corr(df_clean['price'])
print(f"Correlation between total_sqft and price: {correlation}")
```

This correlation indicates a moderate positive relationship. This is intuitive as we expect larger houses to be more expensive in most places.

```python
plt.scatter(df_clean['total_sqft'], df_clean['price'])
plt.xlabel('Total Square Feet')
plt.ylabel('Price')
plt.title('Relationship between Total Square Feet and Price')
plt.show()
```

```python
# regression analysis
X = df_clean['total_sqft']
y = df_clean['price']
X = sm.add_constant(X) 

model = sm.OLS(y, X).fit()
print(model.summary())
```

##### - The r squared value observed here is 0.307, this means that approximately 30.7% of the variation in house prices can be explained by the total square footage of the house
##### - The F statistic is large at 7242 and the asociated p value is zero meaning we can safely reject the null hypothesis and conclude that the total_sqft variable is statistically significant in the model.
##### - The condition number is large and this is indicative of the multicollinearity in the variables incorporated into the new feature.
##### -The Skew is 1.343 which means there could be right skew , while the kurtosis is high indicating a presence of outliers.
##### -The omnibus test value is 4660.339 and the associated p value is 0 suggesting that the residuals are not normally distributed. 

These results communicate that while the model is significant and explains about 30.7% of the variation in prices, it is not perfect due to the lack of a normal distribution of residuals and the high multicollinearity as evidenced by the large condition number

```python
# Fitting some linear regression models 
X_grade = df_clean[['grade']]
y = df_clean['price']
X_grade = sm.add_constant(X_grade)
model_grade = sm.OLS(y, X_grade).fit()
print("Model: Grade")
print(model_grade.summary())
```

##### - The R- sqaured value is telling us that 33.7% of the variation in the house prices can be explained by the grade of the house. 
##### -The coefficient relays that for each unit increase in grade, we can expect the price of the house to increase by about 144,400 dollars on average assuming all other variables are held constant.
##### -The P value is less than 0.05 which means we can reject the null hypothesis and conclude that grade is a statistically significant predictor of price.
##### -The p value of the f statistic is 0 indicating that the model is statistically significant.

```python
#Let's observe the relationship of this variable to price on a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df_clean['grade'], df_clean['price'])
plt.xlabel('Grade')
plt.ylabel('Price')
plt.title('Grade vs Price')
plt.show()
```

##### A lot of vertical scatter typically indicates that the relationship between variables is not perfectly linear or maybe that there is a lot of variability in the data.
##### Since the goal of regression analysis is not to perfectly predict the price but to understand the relationship between these two variables, this information can still be valuable for the insurance company. 
##### The grade of a house is a measure of its quality and higher quality houses are usually expected to have higher prices. This is consistent with what the coefficient is showing. 
##### To confirm, we can tell from the scatter plot above, higher grade houses tend to be more expensive than lower grade houses. This is valuable information for an insurance company looking to estimate total payout in the case of a disaster. They can assume that higher grade houses will be more expensive and thus will require a higher payout.

##### - The R- sqaured communicates that 18.7% of the variation in the house prices can be explained by the square footage of the house apart from the basement. It is a very small value indicating that it doesn't explain the variation of factors too well.
##### -The coefficient relays that for each unit increase in sqft_above, we can expect the price of the house to increase by about 167.47 dollars on average assuming all other variables are held constant.

### Multiple Linear Regression

```python
# Defining independent and dependent variables
X = df_clean[['grade', 'view', 'bedrooms']]
y = df_clean['price']

# Adding a constant to the independent value
X1 = sm.add_constant(X)

# Making the regression model 
model = sm.OLS(y, X1)

# Fitting the  model
results = model.fit()
print(results.summary())

```

##### -The results of this iteration of the multiple linear regression model show that the variables 'grade', 'view', and 'bedrooms' are statistically significant predictors of the price of a house.

##### -For each unit increase in grade, the price of the house increases by approximately 126,200 dollars, holding all other variables constant. This makes sense as the grade of a house is a measure of its quality, and higher quality houses are expected to have higher prices.

##### -For each unit increase in view, the price of the house increases by approximately 87,160 dollars, holding all other variables constant. This suggests that houses with better views are priced higher, which aligns with our intuition.

##### -For each additional bedroom, the price of the house increases by approximately 18,030 dollars, holding all other variables constant. This indicates that larger houses (with more bedrooms) tend to be more expensive.This is also likey why there is strong multicollinearity with bedrooms and total_sqft. As the sqaure footage of the house increases, we would also expect that the number of bedrooms would increase and what would follow is an increase in the price of the house. The insurance company can use this model to look at how the number of bedrooms in a home would affect the total payout in the event of an incident. 

##### -The model has an R-squared value of 0.356, which means that approximately 35.6 percent of the variation in house prices can be explained by these three variables. While this isn't a particularly high R-squared value, we can see that each of these variables affect the price of houses in a significant way. This is definitely a model that the insurance company can use to predict payout prices of houses based on these features. 

```python
# Defining independent and dependent variables
X = df_clean[['condition','total_sqft','view','grade']]
y = df_clean['price']

# Adding a constant to the independent value
X1 = sm.add_constant(X)

# Making the regression model 
model = sm.OLS(y, X1)

# Fitting the  model
results = model.fit()
print(results.summary())
```

##### The R-squared value is 0.418, which means that about 41.8% of the variation in house prices can be explained by the condition, total square footage, view, and grade of the house. 
##### The F-statistic is 2775 and the associated p-value is 0.00 which means that at least one of the independent variables is statistically significant in the model.
##### The condition number is large (3.05e+04), indicating that there might be strong multicollinearity or other numerical problems. 
##### According to the omibus test, the residuals are not normally distributed.
##### Our model is not perfect, due to the non normal distribution and strong multicollinearity. However, it is still useful as it informs that:
    -For every one unit increase in the condition of the house, we can expect the price of the house to increase by about 49,010 dollars, all else being equal.This means that houses in better condition tend to be more expensive. 
    -As for the view, every increase in unit in the quality of the view of the house, the price increases by about 75,480 dollars assuming all other variables are held constant. This is what we would expect intuitively.
    -As for the grade, every increase in a unit in the grade of the house increases the sale price by about 97,630 dollars assuming all other variables are held constant and that is quite a significant amount.
##### This is not a perfect model, but it can definitely be useful in predicting which variables to pay attention to when calculating total payouts. Also, out of all the models ran, it explains the biggest variation in house prices at 41.8%
